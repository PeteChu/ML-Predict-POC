```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./dataset/kc_house_data (Decision Tree).xlsx')
df
```

```python
Y = df.price_range
Y
```

```python
X = df.drop(['date', 'price', 'id', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'price_range'], axis=1)
X
```

```python
target_name = list(set(Y))
feature_name = list(X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(max_depth=5)
```

```python
clf = model.fit(X_train, Y_train)
clf
```

```python
Y_pred = clf.predict(X_test)
Y_pred
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=feature_name, class_names=target_name, filled=True, rounded=True)
graph = graphviz.Source(dot_data)
graph.render('house price (DT)')
```
